```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
```

```python
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies_new=movies.merge(credits, on='title')
```

# **4809 * 23 after merging**

```python
movies_new.columns
```

```python
movies_new1=movies_new[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies_new1.head()
```

```python
movies_new1.isnull().sum()
```

```python
movies_new1.dropna(inplace=True)
```

```python
print(movies_new1.duplicated().sum())
```

```python
movies_new1.head()
```

```python
movies_new1.loc[0,'genres']
```

```python
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies_new1['genres'] = movies_new1['genres'].apply(convert)
```

```python
display(movies_new1.head())
```

```python
movies_new1['keywords'] = movies_new1['keywords'].apply(convert)
```

```python
display(movies_new1.head())
```

```python
movies_new1['cast'] = movies_new1['cast'].apply(convert)
```

```python
display(movies_new1.head())
```

```python
movies_new1['crew'] = movies_new1['crew'].apply(convert)
```

```python
display(movies_new1.head())
```

```python
movies_new1['overview'] = movies_new1['overview'].apply(lambda x:x.split())
```

```python
movies_new1.head()
```

```python
def remove_space(obj):
    L = []
    for i in obj:
        L.append(i.replace(" ",""))
    return L
```

```python
movies_new1['cast'] = movies_new1['cast'].apply(remove_space)
movies_new1['crew'] = movies_new1['crew'].apply(remove_space)
movies_new1['genres'] = movies_new1['genres'].apply(remove_space)
movies_new1['keywords'] = movies_new1['keywords'].apply(remove_space)
```

```python
movies_new1.head()
```

```python
movies_new1['tags'] = movies_new1['overview'] + movies_new1['genres'] + movies_new1['keywords'] + movies_new1['cast'] + movies_new1['crew']
```

```python
movies_new1.head()
```

```python
new_df=movies_new1[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df.iloc[0]['tags']
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

# **Now generating embeddings**

```python
import nltk
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
def stemmer(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stemmer)
```

```python
new_df.iloc[0]['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(new_df['tags']).toarray()
vector
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
  index=new_df[new_df['title']==movie].index[0]
  distances=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
  for i in distances[1:6]:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('American Beauty')
```
